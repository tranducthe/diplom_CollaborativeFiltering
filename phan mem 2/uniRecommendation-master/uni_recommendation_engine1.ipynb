```python
import pandas as pd
```

```python
df1=pd.read_csv(r"F:\LUAN VAN\Bài làm Luận văn\HE THONG GOI Y\csdl\list_uni\chinh\ListUni.csv")
df1.head()
```

```python
df2=pd.read_csv(r"F:\LUAN VAN\Bài làm Luận văn\HE THONG GOI Y\csdl\list_uni\chinh\ratingUni.csv")
df2.head()
```

```python
df3=pd.merge(df1,df2).drop(['Main_topic'],axis=1)
df3.head()
```

```python
userRatings = df3.pivot_table(index=['UserID'],columns=['Name_Uni'],values='Rating')
userRatings.head()
print("before: ",userRatings.shape)

userRatings=userRatings.dropna(thresh=10,axis=1).fillna(0,axis=1)

print("After: ",userRatings.shape)
userRatings.head()
```

```python
corrMatrix = userRatings.corr(method='pearson')
corrMatrix
```

```python
for i in corrMatrix.index:
    print(i)
```

```python
def get_similar(Name_Uni,Ratings):
    similar_ratings=corrMatrix[Name_Uni]*(Ratings-2.5)
    similar_ratings=similar_ratings.sort_values(ascending=False)
    return similar_ratings
```
